# **Data Visualization**

```python
data["published_timestamp"]=pd.to_datetime(data["published_timestamp"])
data["Year"]=data["published_timestamp"].dt.year
year = data.groupby(["Year"])["course_id"].count()
year = pd.DataFrame(year)
year.rename(columns = {"course_id":"Total_Courses"},inplace = True)
year.reset_index(inplace=True)
year
```

```python
sns.barplot(x="Year",y="Total_Courses",data = year)
plt.show()
```

```python
columns1 = ['level','subject']
for i in columns1:
    plt.title(i)
    sns.countplot(data = data, y =i)
    plt.show()
```

```python
sns.boxplot(x = "level", y = "price", data = data)
```

```python
plt.figure(figsize=(7,7))
subject_price_level = sns.barplot(x='subject',y='price',data = data ,hue = 'level')
```

```python
sns.histplot(data=data,x="price",bins=20)
```

```python
sns.histplot(x="price",data = data,kde = True)
```

```python
sns.histplot(data=data,x="num_subscribers",bins=10);
```

```python
top_paid_courses = data[data["price"]!=0][["course_title","num_subscribers","subject"]].sort_values(by = "num_subscribers",ascending = False)
top_paid_courses.head(3)
```

```python
top_unpaid_courses = data[data["price"]==0][["course_title","num_subscribers","subject"]].sort_values(by = "num_subscribers",ascending = False)
top_unpaid_courses.head(3)
```

```python
plt.title("Paid Courses")
sns.countplot(y ="subject", data = top_paid_courses);
plt.show()
plt.title("UnPaid Courses ")
sns.countplot(y ="subject", data = top_unpaid_courses);
plt.show()
```

```python
data_free_course = data.loc[data.is_paid == False]
data_free_course.head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
scatter = sns.scatterplot(x='num_lectures',y='num_subscribers',data=data)
```

```python
#5 Most Popular Courses/num_subscribers
top_courses = data.sort_values(by='num_subscribers', ascending=False)[:5]
figure = plt.figure(figsize=(10,10))
sns.barplot(y=top_courses.course_title, x=top_courses.num_subscribers)
plt.xticks()
plt.xlabel('number of subscribers')
plt.ylabel('Course title')
plt.title('The Most Popular courses')
plt.show()
```

```python
#Top 5 Num_lectures by num_lectures
top_course = data.sort_values(by='num_lectures', ascending=False)[:5]
figure = plt.figure(figsize=(5,5))
sns.barplot(y=top_course.course_title, x=top_course.num_lectures)
plt.xticks()
plt.xlabel('num_lectures')
plt.ylabel('Course_title')
plt.title('Number of lectures Courses')
plt.show()
```
